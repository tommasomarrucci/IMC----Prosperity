```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv("Dati.csv", delimiter=";")

# Filter the data
kelp_df = df[df['product'] == 'KELP']
squid_ink_df = df[df['product'] == 'SQUID_INK']
rainforest_resin_df = df[df['product'] == 'RAINFOREST_RESIN']

# Create subplots WITHOUT sharey
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Plot KELP
axs[0].plot(kelp_df['timestamp'], kelp_df['mid_price'], linewidth = 0.5)
axs[0].set_title('KELP')
axs[0].set_xlabel('Timestamp')
axs[0].set_ylabel('Mid Price')
axs[0].grid(True)

# Plot SQUID INK
axs[1].plot(squid_ink_df['timestamp'], squid_ink_df['mid_price'])
axs[1].set_title('SQUID INK')
axs[1].set_xlabel('Timestamp')
axs[1].grid(True)

# Plot RAINFOREST RESIN
axs[2].plot(rainforest_resin_df['timestamp'], rainforest_resin_df['mid_price'], linewidth = 0.5)
axs[2].set_title('RAINFOREST RESIN')
axs[2].set_xlabel('Timestamp')
axs[2].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```
