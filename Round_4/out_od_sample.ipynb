```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# 1) Read and extract the “Activities log” section
file_path = '/Users/tommasomarrucci/Desktop/IMC BIEF 42/3.log'
with open(file_path, 'r') as f:
    lines = f.readlines()

start_marker = "Activities log:"
header = None
data_lines = []
in_section = False
for line in lines:
    if start_marker in line:
        in_section = True
        continue
    if in_section:
        if line.strip() == "" or line.strip().startswith("{"):
            break
        if header is None:
            header = line.strip()
        else:
            data_lines.append(line.strip())

csv_text = "\n".join([header] + data_lines)
df = pd.read_csv(StringIO(csv_text), sep=';')

# 2) Pivot to get mid_price series per product
pivot = df.pivot(index='timestamp', columns='product', values='mid_price')

# 3) (Optional) convert timestamp to datetime if needed
# base_date = pd.to_datetime('2025-04-03')
# pivot.index = base_date + pd.to_timedelta(pivot.index, unit='s')

# 4) Loop over products and plot each in its own figure
for product in pivot.columns:
    plt.figure(figsize=(8,4))
    plt.plot(pivot.index, pivot[product], linewidth=1)
    plt.title(f"Mid‑Price of {product} Over Time")
    plt.xlabel("Timestamp")
    plt.ylabel("Mid Price")
    plt.grid(True)
    plt.tight_layout()
    # Uncomment next line to save each plot as a PNG:
    # plt.savefig(f"mid_price_{product}.png", dpi=150)
    plt.show()
```
